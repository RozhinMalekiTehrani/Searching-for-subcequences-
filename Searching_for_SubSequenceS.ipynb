```python

def find_pattern(data, chromosome, patterns):
            # ch = 'CP052038.1'
#             sequence = data[chromosome]
            positions = []
            START = []

        # we consider all the strings are capitall
            sequence = MutableSeq(data[chromosome]).upper()





            END = []
            pat = []
            for pattern in patterns:

                for match in re.finditer(pattern, str(sequence)):
                    start, end = match.span()

                    is_overlapping = any(pos_start <= start < pos_end or pos_start < end <= pos_end for pos_start, pos_end in positions)
            #         print(is_overlapping, pattern)

                    if not is_overlapping:
                        positions.append((start, end))

                        START.append(start+1)
                        END.append(end)
            #             p = pattern* len(is_overlapping)
                        pat.append(pattern)


            CHR = [chromosome]* len(pat)


            OUTPUT = {'CHR':CHR,'PATTERN':pat ,'POSITION_start':START,'POSITION_end':END }#, 'LABEL':[]}

            Result = pd.DataFrame(OUTPUT)
            Result = Result.sort_values(by = ['POSITION_start'])
            return(Result)
            # positions.sort()
            # print(positions)
```
